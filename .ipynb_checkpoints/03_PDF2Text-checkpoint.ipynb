```python
import os
import time
import numpy as np
from PyPDF2 import PdfReader
```

# `from PyPDF2 import PdfReader`

### `0.05 seconds per page`
### `-> 100 Million pages per week with 10 parallel processes`


## Example 1


```python
# Example 1
reader = PdfReader("./data/example_1.pdf")
n_pages = len(reader.pages)

dur1 = time.time()
txtList = []
for i in range(n_pages):
    page = reader.pages[i]
    txtList.append(page.extract_text())
dur1 = (time.time() - dur1) / n_pages
print(txtList[0])
```

## Example 2



```python
# Example 2
reader = PdfReader("./data/example_2.pdf")
n_pages = len(reader.pages)

txtList = []
dur2 = time.time()
for i in range(n_pages):
    page = reader.pages[i]
    txtList.append(page.extract_text())
dur2 = (time.time() - dur2) / n_pages
print(txtList[1])
```

```python
# Example 3
reader = PdfReader("./data/example_3.pdf")
n_pages = len(reader.pages)

dur3 = time.time()
txtList = []
for i in range(n_pages):
    page = reader.pages[i]
    txtList.append(page.extract_text())
dur3 = (time.time() - dur3) / n_pages
print(txtList[0])
```
