```python
from matplotlib import pyplot as plt
from scipy.stats import halfcauchy, norm, invgamma
import numpy as np
```

```python
x_grid = np.linspace(0,6,10000)

yC = halfcauchy.pdf(x=x_grid, loc=0, scale=1)
yN = 2.0 * norm.pdf(x=x_grid, loc=0, scale=1)
```

```python
plt.figure(figsize=(10,7))
plt.plot(x_grid, yN, color='#48b4e0')
plt.plot(x_grid, yC, color='#D2042D')
plt.xlim(-0.2,6)
plt.xlabel('x', fontsize=14)
plt.ylabel('$f(x)$', fontsize=14)
plt.legend({r'$\mathcal{N}(0,1)\,|_{\mathbb{R}_{\geq 0}}$' : 'blue',  r'$\mathcal{C}_{+}(0,1)$' : 'red'}, fontsize=12)
```

```python
ig1 = invgamma.pdf(x_grid, 1, scale=1)
ig2 = invgamma.pdf(x_grid, 1, scale=2)
ig3 = invgamma.pdf(x_grid, 1, scale=3)
#yN = 2.0 * norm.pdf(x=x_grid, loc=0, scale=1)
```

```python
#plt.figure(figsize=(10,7))
plt.plot(x_grid, ig1, color='red')
plt.plot(x_grid, ig2, color='#0000a5')
plt.plot(x_grid, ig3, color='green')
plt.xlabel('x', fontsize=14)
plt.ylabel('$f(x)$', fontsize=14)
plt.legend({r'$\mathcal{IG}(1,1)$' : 'blue',  r'$\mathcal{IG}(1,2)$' : '#0000a5', r'$\mathcal{IG}(1,3)$' : 'green'}, fontsize=12)
#plt.plot(x_grid, yC, color='#D2042D')
#plt.xlim(-0.2,6)
#plt.xlabel('x', fontsize=14)
#plt.ylabel('$f(x)$', fontsize=14)
#plt.legend({r'$\mathcal{N}(0,1)\,|_{\mathbb{R}_{\geq 0}}$' : 'blue',  r'$\mathcal{C}_{+}(0,1)$' : 'red'}, fontsize=12)
```

```python
ig1 = np.mean(invgamma.rvs(1, scale=1, size=100000))
ig2 = np.mean(invgamma.rvs(1, scale=2, size=100000))
ig3 = np.mean(invgamma.rvs(1, scale=3, size=100000))
```
