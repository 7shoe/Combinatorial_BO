## 3. Graph Fourier Transform (GFT) `???`

#### Init graph and run `graph fourier transform`

```python
# init (arbitrary) networkx graph
G = nx.Graph(); G.add_edge(0, 1); G.add_edge(1, 2); G.add_edge(2, 3); G.add_edge(3, 4); G.add_edge(4, 0)

# comput weight(ed adjacency) matrix W
W = np.zeros(shape=(len(G), len(G)))
rows, cols = zip(*list(G.edges))
W[rows, cols] = 1
W+=W.T

# init PyGSP graph (from W)
G = graphs.Graph(W)

# compute Graph Fourier Transform (GTF)
G.compute_fourier_basis()

# apply GFT
s_1 = np.array([0.5, 4.6, 5.6, 3.56, 0.5])

gft_1 = G.gft(s=s_1)
```
