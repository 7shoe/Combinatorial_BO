```python
%%time 

# set alpha
n = 10
sigma = 0.2

N_EXP = 30

# alpha
alpha = np.array([1.7, -0.,  1.3, -0.3, -0.5,  1. ,  3. ,  0.4,  1.1,  2.4,  0.2])
a     = np.array([0] + list(alpha) + [0]*math.comb(11,2))

# 
mseDict = {}

# loop experiments
for rep in range(N_EXP):
    newList = []
    for n in range(10,300,2):
        # sample data
        k = len(alpha)
        X = np.random.binomial(n=1, p=0.5, size=n*k).reshape(n,-1)
        y = X @ alpha.T + np.random.normal(0,sigma,size=n)

        # Bayes reg
        sbr1 = SparseBayesReg(N_total=600, burnin=100, thinning=5)
        sbr1.fit(X,y)
        
        # measure l2
        mse_loc = np.linalg.norm(a-sbr1.getMeanAlpha(), ord=2)
        
        # append
        newList.append(mse_loc)
        
    # append
    mseDict[rep] = newList
    
# store
with open('./data/new_mseDict.json', 'w') as f:
    json.dump(mseDict, f)
```

```python
new_mseDict = {}
for k in mseDict.keys():
    new_mseDict[k] = np.sqrt(np.array(mseDict[k]) - 0.2)
```
